## Embedding model used in this example : sentence-transformers/all-MiniLM-L6-v2


This is a sentence-transformers model: It maps sentences & paragraphs to a 384 dimensional dense vector space and can be used for tasks like clustering or semantic search.

```python
from langchain_huggingface import HuggingFaceEmbeddings
```

```python
embeddings = HuggingFaceEmbeddings(model_name='sentence-transformers/all-MiniLM-L6-v2')
```

```python
query = "I am a tech savy and love travelling"
```

```python
vectors = embeddings.embed_query(query)
```

```python
len(vectors)
```

```python
documents = [
    "Sachin Tendulkar is often called the 'God of Cricket'. He was the first player to score 100 international centuries",
    "Virat Kohli is known for his passion and consistency, he has been a run-machine across formats and a former India captain.",
    "Rohit Sharma is the current Indian captain, famous for his elegant stroke play and three ODI double centuries",
    "Jasprit Bumrah is India’s pace spearhead, known for his deadly yorkers and unorthodox bowling action.",
    "Ravindra Jadeja is a dynamic all-rounder, Jadeja contributes with bat, ball, and exceptional fielding."
]
```

```python
query = "Tell me about Virat Query"
```

```python
doc_embeddings = embeddings.embed_documents(documents)
query_embedding = embeddings.embed_query(query)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_scores = cosine_similarity([query_embedding], doc_embeddings)[0]
enum_similarity_scores = list(enumerate(similarity_scores))
sorted_similarity_scores = sorted(enum_similarity_scores, key=lambda x: x[1])
sorted_similarity_scores
```

```python
index, doc_score = sorted_similarity_scores[-1]
print("Answer to query: \n", documents[index])
```
